```python
#Pedimos o importamos dos cosas de esta libreria. El beautiful soup que extrae del sitio web,y una serie de
#solicitudes o requests genericos que funciona con beutiful soup.
from bs4 import BeautifulSoup
import requests 
```

```python
#¿que url queres solicitar? La cargamos. También nuestro header que no siempre es necesario.
url = 'https://www.op.gg/leaderboards/tier?region=kr&tier=bronze&page=1'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36 Edg/121.0.0.0'
}
```

```python
#La solicitud o nombre del request con la que obtendriamos lo que cargamos como url. 
#Ádemas esto que obtenemos lo colocamos en page. 
page = requests.get(url,headers=headers)
```

```python
#Ponemos la acción del beautifull, que solicita page en formato texto, en soup para imprimir.
soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
#se busca la tabla que este en la pagina. Si se diera el caso que la pagina tuviera diferentes tablas se las extrae a todas.
#En tal caso se puede unar un Index para buscar la que queremos, o se la puede buscar con el nombre especifico. 
soup.find('table')
#Una forma alternativa de buscar la tabla de una manera más especifica. Quizas sirva para buscar atributos especificos:
#soup.find('table', class_ = 'css-44pun euud7vz10')
```

```python
#Trabajamos con la tabla que nos interesa. Ponemos la operación en 'tabla' y dejamos de operar en la soup.
table = soup.find('table')
```

```python
#empezamos a trabajar con los headers que nos interesan. Primero los ubicamos en la tabla (ya no en la soup). Si hubieran más 
#tablas nos pasaria que extraeria th que no precisamos.
table.find_all('th')
```

```python
#le asignamos un nombre.
leaderboard_titles = table.find_all('th')
#verificamos y vemos.
leaderboard_titles
```

```python
#Solo dejamos el text, es una limpieza. En cada string hay un loop que solo deja el texto.
#el .strip no es necesario en este caso. Pero es util para quitar elementos extras y tener un resultado más limpio.
leaderboard_table_titles = [title.text.strip() for title in leaderboard_titles]
#resultado. 
print(leaderboard_table_titles)
```

```python
import pandas as pd
```

```python
#introducimos los data frame para poder trabajar con tablas, es decir, ordenar datos en columnas y filas.
df = pd.DataFrame(columns = leaderboard_table_titles) 

df
```

```python
#empezamos a trabajar con la row 
column_data = table.find_all('tr')
```

```python
#El problema es que son lineas independientes y ahora hay que meter todo en una tabla, en la data frame.
#Se hizo un pequeño arreglo a la primera string para que se saltee la primera linea que estaba vacia ocacionando un error.
for row in column_data[1:]: 
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length]= individual_row_data
```
